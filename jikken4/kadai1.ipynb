```python
import numpy as np
import matplotlib.pyplot as plt

# df = pd.read_csv('test.txt', index_col=None, index_row=None)
data = np.genfromtxt('test.txt', delimiter=' ')

sorted_indices = np.argsort(data[:,0])
data_sorted = data[sorted_indices]

plt.figure()
plt.plot(data_sorted[:, 0], data_sorted[:, 1])
plt.title('Test data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
