```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

for a in np.arange(0.01,0.51, 0.01):
    data = np.genfromtxt(f'./kadai2_data/train_{int(a*100):02d}.txt', delimiter=' ')
    sorted_indices = np.argsort(data[:,0])
    data_sorted = data[sorted_indices]
    plt.figure()
    plt.plot(data_sorted[:, 0], data_sorted[:, 1])
    plt.title(f'Plot for a = {a:.2f}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```
