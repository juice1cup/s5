```python
import numpy as np
import matplotlib.pyplot as plt
import os

data = np.genfromtxt('./test.txt', delimiter=' ')

folder_name = 'kadai2_data'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
np.random.seed(42)
indices = np.random.choice(len(data), size = len(data)//2, replace=False)
selected_data = data[indices]

for a in np.arange(0.01,0.51, 0.01):
    filename = f'train_{int(a*100):02d}.txt'
    filepath = os.path.join(folder_name, filename)
    noisy_data = selected_data + np.random.normal(0, a, selected_data.shape)
    np.savetxt(filepath, noisy_data, fmt='%.6e')

sorted_indices = np.argsort(data[:,0])
data_sorted = data[sorted_indices]

plt.plot(data_sorted[:, 0], data_sorted[:, 1])
```
